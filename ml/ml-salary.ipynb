```python
import pybaseball as ball
import pandas as pd
import numpy as np  
import matplotlib.pyplot as plt  
import seaborn as seabornInstance 
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression
from sklearn import metrics
from sklearn.impute import SimpleImputer
%matplotlib inline
```

## Data Cleaning and Dataframe Creation

```python
salaries = []
years = ['19','18','17','16','15','14']
```

```python
for year in years:
    url = 'https://www.usatoday.com/sports/mlb/salaries/' + year + '/player/all/'
    df = pd.read_html(url)
    salaries.append(df[0])
```

```python
batters19 = pd.read_csv('stat_files/b19.csv')
batters18 = pd.read_csv('stat_files/b18.csv')
batters17 = pd.read_csv('stat_files/b17.csv')
batters16 = pd.read_csv('stat_files/b16.csv')
batters15 = pd.read_csv('stat_files/b15.csv')
batters14 = pd.read_csv('stat_files/b14.csv')
pitchers19 = pd.read_csv('stat_files/p19.csv')
pitchers18 = pd.read_csv('stat_files/p18.csv')
pitchers17 = pd.read_csv('stat_files/p17.csv')
pitchers16 = pd.read_csv('stat_files/p16.csv')
pitchers15 = pd.read_csv('stat_files/p15.csv')
pitchers14 = pd.read_csv('stat_files/p14.csv')
```

```python
batting_salaries = []
pitching_salaries = []
    
```

```python
for i in salaries:
    df = i[['Name', 'POS', 'Avg Annual']]
    df.dropna()
    batting_salaries.append(df[(df['POS'] != 'SP') & (df['POS'] != 'RP')])
    pitching_salaries.append(df[(df['POS'] == 'SP') | (df['POS'] == 'RP')])
```

```python
batting_stats = [batters19,batters18,batters17,batters16,batters15,batters14]
pitching_stats = [pitchers19,pitchers18,pitchers17,pitchers16,pitchers15,pitchers14]
```

```python
combined_bats = []
combined_pitch = []
```

```python
for i in range(6):
    stats = batting_stats[i]
    salary = batting_salaries[i]
    combine = pd.merge(stats, salary, how='left', left_on=['Player'], right_on = ['Name'])
    combined_bats.append(combine)
```

```python
for i in range(6):
    stats = pitching_stats[i]
    salary = pitching_salaries[i]
    combine = pd.merge(stats, salary, how='left', left_on=['Player'], right_on = ['Name'])
    combined_pitch.append(combine)
```

```python
final_bat = pd.concat(combined_bats)
final_pitch = pd.concat(combined_pitch)
```

```python
final_bat['Avg Annual'] = final_bat[final_bat.columns[-1]].replace('[\$,]', '', regex=True).astype(float)
final_pitch['Avg Annual'] = final_pitch[final_pitch.columns[-1]].replace('[\$,]', '', regex=True).astype(float)
final_bat.dropna()
final_pitch.dropna()
final_pitch.head(3)
```

```python
# #filter out pitchers who have pitched less than 30 innings
# pfilter = final_pitch['IP'] > 30
# fpitch = final_pitch[pfilter]
# #filter out batters with less than 50 at bats
# bfilter = final_bat['AB'] > 50
# fbat = final_bat[bfilter]
```

```python
fbat = final_bat.drop(['Name', 'POS'], axis = 1) 
fpitch = final_pitch.drop(['Name', 'POS'], axis = 1) 
```

```python
fbat = fbat.dropna()
fpitch = fpitch.dropna()
```

## Batter Training

```python
#creatings a veterans DF with players making over 5M
vets = fbat['Avg Annual'] >500000
vets = fbat[vets]
vets.head(3)
```

```python
y=fbat['Avg Annual']
# X=fbat.iloc[:,3:-1]
X=fbat[['AB','H','HR','BB','SO','AVG','OBP','OPS']]
```

```python
print(X.shape)
print(y.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=0)
```

```python
regressor = LinearRegression()  
regressor.fit(X_train, y_train)
```

```python
coeff_df = pd.DataFrame(regressor.coef_, X.columns, columns=['Coefficient'])  
coeff_df
```

```python
y_pred = regressor.predict(X_test)
```

```python
df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
df
```

```python
df1 = df.head(25)
df1.plot(kind='bar',figsize=(16,10))
plt.grid(which='major', linestyle='-', linewidth='0.5', color='green')
plt.grid(which='minor', linestyle=':', linewidth='0.5', color='black')
plt.show()
```

```python
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))  
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))  
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

```python
np.sqrt(metrics.mean_squared_error(y_test, y_pred))/fbat['Avg Annual'].mean()
```
