```python
salaries = []
years = ['2019','2018','2017','2016','2015','2014']
for year in years:
    url = 'https://www.usatoday.com/sports/mlb/salaries/' + year + '/player/all/'
    df = pd.read_html(url)
    df = df [0]
    df['year']= year
    salaries.append(df)
```

```python
for year in salaries:
    for index, row in year.iterrows():
        salary = {row['year']:row['Salary']}
        query = {'playerName':row['Name']}
        update = {'$set': {'Salary':salary}}
        print(query, update)
        try:
            collection.update_one(query, update)
        except:
            pass
```

```python
names = [i['playerName'] for i in collection.find()]
names
```

```python
for player in names:
    s = {}
    for df in salaries:
        try:
            row = df.loc[df['Name'] == player]
            avg_salary = int(row['Avg Annual'].tolist()[0][1:].replace(',',''))
            year = row['year'].tolist()[0]
            s[year] = avg_salary
        except:
            pass
    x = {'Salary':s}
    query = {"playerName":player}
    update = {"$set": x}
    collection.update_one(query, update)
```
